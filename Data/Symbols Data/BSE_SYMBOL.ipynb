```python
import pandas as pd
import duckdb
import os

# Path to your bhavcopy CSV
csv_path = r"C:\Users\jallu\OneDrive\pgp\Python\Stock predictor\Rubik_view\Data\Symbols Data\bhavcopy.csv"

# Path to your DuckDB database
db_dir = r"C:\Users\jallu\OneDrive\pgp\Python\Stock predictor\Rubik_view\Data\Symbols Data"
os.makedirs(db_dir, exist_ok=True)
db_path = os.path.join(db_dir, "symbols.duckdb")

# Load the CSV into a DataFrame
df_bse_meta = pd.read_csv(csv_path)
df_bse_meta.columns = df_bse_meta.columns.str.strip().str.upper().str.replace(" ", "_")

# Add 'MISSING_DATA' column
df_bse_meta['MISSING_DATA'] = df_bse_meta.isnull().any(axis=1).map({True: 'Yes', False: 'No'})

# Save DataFrame to DuckDB as a new table 'bse_metadata'
con = duckdb.connect(db_path)
con.execute("DROP TABLE IF EXISTS bse_metadata")
con.execute("CREATE TABLE bse_metadata AS SELECT * FROM df_bse_meta")
print(f"\n✅ All columns from bhavcopy.csv imported as table 'bse_metadata' in {db_path}")
print("\nSample rows:")
print(con.execute("SELECT * FROM bse_metadata LIMIT 5").fetchdf())

# Show all tables in the DB (optional)
tables = con.execute("SHOW TABLES").fetchdf()
print("\nTables in DB now:")
print(tables)

con.close()
```
